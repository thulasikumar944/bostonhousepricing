```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
```

```python
df=load_boston()
```

```python
data=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
data.head()
```

```python
data['Price']=df.target
```

```python
#check missing values
data.isnull().sum()
```

```python
data.describe()
```

```python
sns.heatmap(data.corr())
```

```python
data.corr()
```

```python
plt.scatter(data['RAD'],data['Price'])
plt.xlabel("Price")
plt.ylabel("Crime Rate")
```

```python
sns.regplot(x="LSTAT",y="Price",data=data)
```

```python
# Making X and Y
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.29,random_state=20)
```

```python
x_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
type(scaler)
```

```python
x_train=scaler.fit_transform(x_train)
```

```python

x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.coef_
```

```python
regressor.get_params()
```

```python
y_pred=regressor.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
```

```python
residulas=y_test-y_pred
```

```python
sns.displot(residulas,kind="kde")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
